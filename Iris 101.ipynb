```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("Keys of iris_datasets: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR']+'\n')
```

```python
print('Target names: {}'.format(iris_dataset['target_names']))
```

```python
print('Feature names: {}'.format(iris_dataset['feature_names']))
```

```python
print('Data type: {}'.format(type(iris_dataset['data'])))
```

```python
print('Shape of Data: {}'.format(iris_dataset['data'].shape))
```

```python
print('Type of target: {}'.format(type(iris_dataset['target'])))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
print('x_train shape: {}'.format((x_train).shape))
```

```python
print('y_train shape: {}'.format((y_train).shape))
```

```python
print('x_test shape: {}'.format((x_test).shape))
print('y_test shape: {}'.format((y_test).shape))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

iris_dataframe = pd.DataFrame(x_train, columns=iris_dataset.feature_names)
```

```python
grr = scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15),marker = 'o',hist_kwds={'bins':20},s=60,alpha=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors =1)
```

```python
knn.fit(x_train,y_train)
```

```python
import numpy as np
x_new = np.array([[5,2.9,1,0.2]])
```

```python
prediction =knn.predict(x_new)
print('Prediction:{}'.format(prediction))
```

```python
y_pred = knn.predict(x_test)
print('Test set predictions:\n{}'.format(y_pred))
```

```python
print('Test set score: {:.2f}'.format(np.mean(y_pred == y_test)))
```

```python
print(print('Test set score: {:.2f}'.format(knn.score(x_test,y_test))))
```
